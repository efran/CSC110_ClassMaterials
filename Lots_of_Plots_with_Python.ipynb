## Bar Plots
Bar plots can be a good choice when you have __counts__ of items.<br><br>
For example, supposed you had a block of text and you counted the occurrences of the words _good_, _bad_, _nice_, _sad_, _happy_, _lol_ (perhaps you were trying some rudimentary sentiment analysis). <br><br>
You could create a bar plot showing the counts you obtained for each word, as shown below.

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
words = ('good', 'bad', 'nice', 'sad', 'happy', 'lol')
counts = [11,8,5,5,2,1]

word_pos = np.arange(0,len(words)) # this sets up the x-axis spacing

 
plt.bar(word_pos, counts, align='center', alpha=0.7) # alpha determines the shading of the bars ("transparency")
# The generates a bar for the counts centered according to the spacing determined for the x axis

plt.xticks(word_pos, words) # This labels the x-axis -- positioning each word bolow the center of its respective bar
plt.ylabel('Word Count') # this labels the y axis

plt.title('Word Counts in Texts')
 
plt.show()

# You can try different alpha values to find a shading you like
```

You can also create a horizontal bar chart, and specify the color you want the bars to be, as shown below:

```python
plt.barh(word_pos, counts, align='center', alpha=0.4, color='green') # the barh specifies a horizontal bar plot
# The color option specifies green

plt.yticks(word_pos, words) # notice this has changed from xticks to yticks
plt.xlabel('Word Count') # notice this has changed from ylabel to xlabel

plt.title('Word Counts in Texts')
 
plt.show()
```

```python
data = [(4,2), (4,10), (7,4), (7,22), (8,16), (9,10), (10,18), (10,26), (10,34), (11,17), (11,28), (12,14), (12,20), 
        (12,24), (12,28), (13,26), (13,34), (13,34), (13,46), (14,26), (14,36), (14,60), (14,80), (15,20), (15,26), 
        (15,54), (16,32), (16,40), (17,32), (17,40), (17,50), (18,42), (18,56), (18,76), (18,84), (19,36), (19,46), 
        (19,68), (20,32), (20,48), (20,52), (20,56), (20,64), (22,66), (23,54), (24,70), (24,92), (24,93), (24,120), 
        (25,85)]

# data obtained from: https://vincentarelbundock.github.io/Rdatasets/datasets.html
```

## Scatter Plots
As mentioned above, specifying a marker shape and __NO__ line style produces a _Scatter Plot_.<br>
This dataset is probably better represented as a scatter plot, rather than a line plot, as the separate pairs of values represent different (i.e. unrelated) cars.

```python
plt.plot(speed, distance, 'g*') # here, "g--" specifies "green, dashed line"

plt.xlabel('speed (mph)')
plt.ylabel('stopping distance (feet)')

# You can make a title for your plot by:
plt.title('Speed and Stopping Distance of Cars (from the 1920s)')

# show the plot on the screen
plt.show()
```

## More on Bar Plots
Sometimes, you might want to use bar plots to show the difference between 2 sets of data.<br>
For example, the following plot shows how much Alice and Bob ate at a recent party (I don't know why you'd want to compare these, but...):


```python
Alice = {"shrimp" : 5, "cookies": 2, "carrot sticks" : 7, "eggs" : 2, "kebab" : 3, "burger" : 0}
Bob = {"shrimp" : 12, "cookies": 4, "carrot sticks" : 0, "eggs" : 4, "kebab" : 3, "burger" : 2}
Carl = {"shrimp" : 1, "cookies": 1, "carrot sticks" : 1, "eggs" : 1, "kebab" : 1, "burger" : 1}

food_items = list(Alice.keys())

a_food = []
b_food = []
c_food = []

for item in food_items:
    a_food.append(Alice[item])
    b_food.append(Bob[item])
    c_food.append(Carl[item])
    
# check your lists:
print(food_items)
print(a_food)
print(b_food)
print(c_food)
```

```python
# data to plot
n_groups = 6 # the number of different foods you are comparing
 
# create plot
fig, ax = plt.subplots()  # this sets up a bar plot with multiple inputs for plotting
index = np.arange(n_groups) # prepare the x axis (for spacing)
bar_width = 0.25
opacity = 0.8
 
a_bars = plt.bar(index, a_food, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Alice')

# notice how the bar width is added to the index -- what happens if you don't do this? 
b_bars = plt.bar(index + bar_width, b_food, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Bob')

c_bars = plt.bar(index + 2*bar_width, c_food, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Carl')
 
plt.xlabel('Food Item')
plt.ylabel('Number Eaten')
plt.title('Party Food Eaten by Alice,Bob, and Carl')
plt.xticks(index + bar_width, food_items)
plt.legend()
 
plt.tight_layout()
plt.show()
```

## More on line plots
Going back to the car speed and stopping distance idea, but using made-up data.<br>
The scenario this time is 4 cars tested at different speeds. The cars are also being drive by different people.

```python
car_1 = {10:27, 15:44, 20:63, 25:85, 30:109, 35:136, 40:164, 45:196, 50:229, 55:265, 60:304 }
car_2 = {10:20, 15:35, 20:55, 25:62, 30:95, 35:102, 40:150, 45:180, 50:212, 55:245, 60:287 }
car_3 = {10:32, 15:49, 20:73, 25:101, 30:120, 35:150, 40:183, 45:226, 50:260, 55:287, 60:348 }
car_4 = {10:26, 15:40, 20:67, 25:83, 30:100, 35:145, 40:170, 45:210, 50:240, 55:301, 60:323 }
```

```python
speeds = sorted(list(car_1.keys()))

driver_1 = []
driver_2 = []
driver_3 = []
driver_4 = []

for speed in speeds:
    driver_1.append(car_1[speed])
    driver_2.append(car_2[speed])
    driver_3.append(car_3[speed])
    driver_4.append(car_4[speed])
    
# Check lists 
print(speeds)
print(driver_1)
print(driver_2)
print(driver_3)
print(driver_4)
```

```python
# Make x, y arrays for each graph
x = speeds
y1 = driver_1
y2 = driver_2
y3 = driver_3
y4 = driver_4

# use plt.plot to plot x and y 
plt.plot(x, y1, 'r--')
plt.plot(x, y2, 'g-.')
plt.plot(x, y3, 'c')
plt.plot(x, y4, 'm:')

# give plot a title
plt.title('Plot of Stopping Distances')

# make axis labels
plt.xlabel("car speed (mph)")
plt.ylabel("stopping distance (feet)")

# set axis limits
plt.xlim(0.0, 70.0)
plt.ylim(0.0, 400.0)

# add a legend
plt.legend(("driver 1", "driver 2", "driver 3", "driver 4"), loc=0)

# loc = 0 means Python will choose the "best" location for the legend

# If you want to specify the location yourself, here are the codes:
# best --> 0
# upper right --> 1 
# upper left --> 2
# lower left --> 3
# lower right --> 4
# right --> 5
# center left --> 6
# center right --> 7
# lower center --> 8
# upper center --> 9
# center --> 10

# show the plot on the screen
plt.show()
```

## Histograms
Histograms can be a handy way to examine a dataset and to see how the data are distributed within the dataset.<br>
To create a dataset to use, we'll combine all the drivers stopping distances into one file.

```python
# combine datasets to create one large file - this has no meaning other than to create a large sample dataset

data = y1 + y2 + y3 + y4 # python makes it easy to combine items in lists into one list
```

```python
# make a histogram of the data array
plt.hist(data, alpha = 0.5, edgecolor="k")
# make plot title & labels
plt.title('Histogram of Value in Dataset')
plt.xlabel("values")
plt.ylabel("count")

plt.show()
```

This is fine, but it is difficult to know what the limits of each bin are.
We can change the formatting to make this histogram easier to read.

```python
# For planning purposes, what are the max and min values in the dataset?
max_val = max(data)
min_val = min(data)

# Before you decide how many bins you want in your histogram, look at how the number of bins fits your range
num_bins = 14
range = max_val - min_val
width_14 = range/num_bins

print("With 14 bins:", width_14)

# If we adjust the range to 330 and make 15 bins, this will fit better
min_val = 20
max_val = 350 # --> this means our histogram will go up to 350
num_bins = 15
range = max_val - min_val

width_15 = range/num_bins
print("With 15 bins:", width_15)
```

```python
# For some reason, the range function stopped working for me, 
# and even restarting the kernel didn't help.
tick_values = list(range(20, 351, 22))
```

```python
# So, in searching for reasons why this might be happening, 
# I landed on this page: http://pythoncentral.io/pythons-range-function-explained/
# and just using their version of range seemed like a good solution.
def getrange(start, stop, step):
    i = start
    while i < stop:
        yield i
        i += step

tick_values = list(getrange(20, 351, 22)) # use 351 so we get the final 350

# this works nicely:
print(tick_values)
```

```python
# make a histogram of the data array
min_val = 20
max_val = 351
val_range = max_val - min_val
num_bins = 14 
bin_width = val_range/num_bins
new_max = min_val+(bin_width*num_bins)+1

tick_values = list(getrange(min_val, new_max, bin_width)) # use 351 so we get the final 350

plt.hist(data, bins=np.arange(min_val, new_max, bin_width),facecolor='blue', edgecolor = "k", alpha=0.3)
# data is the list of data values to make the histogram from
# bins uses the same values we used to create the tick_values list
# facecolor and opacity(alpha) set column appearance
# edgecolor = "k" outlines the columns in black

# make plot title & labels
plt.title('Histogram of Values in Dataset')
plt.xlabel("values")

plt.xticks(tick_values) # make upper limit 351 since range doesn't include upper bound
# range(lowest, highest+1, step value) 
# step value = bin width --> this ensures the bins are of equal width
# bin width = (highest - lowest) / number of bins --> since this might not be an integer value - cast it to an int
plt.ylabel("count") 

plt.show()

# Note: The histogram is showing how many values in the dataset fall in each bin
# for example, 8 values in the dataset are between 20 and 53

```

Hmm... We're having a problem with our bin width not being an integer.<br>
We would prefer the x-axis ticks to be located on integer values.<br>
Let's try to fix this.

```python
# make a histogram of the data array
min_val = 20
max_val = 350
val_range = max_val - min_val
num_bins = 14 
bin_width = val_range/num_bins
new_max = min_val+(bin_width*num_bins)+1 # Add one because range doesn't include upper value

# Now, to get rid of the decimals, we can round every value in the list.
tick_values_unrounded = list(getrange(min_val, new_max, bin_width))
tick_values = []
for val in tick_values_unrounded:
    tick_values.append(round(val))

plt.hist(data, bins=np.arange(min_val, new_max, bin_width),facecolor='blue', edgecolor = "k", alpha=0.3)
# data is the list of data values to make the histogram from
# bins uses the same values we used to create the tick_values list
# facecolor and opacity(alpha) set column appearance
# edgecolor = "k" outlines the columns in black

# make plot title & labels
plt.title('Histogram of Values in Dataset')
plt.xlabel("values")

plt.xticks(tick_values) # make upper limit 351 since range doesn't include upper bound
# range(lowest, highest+1, step value) 
# step value = bin width --> this ensures the bins are of equal width
# bin width = (highest - lowest) / number of bins --> since this might not be an integer value - cast it to an int
plt.ylabel("count") 

plt.show()

# Note: The histogram is showing how many values in the dataset fall in each bin
# for example, 6 values in the dataset are fall within the range represented by the first bar
```

This is much better, but if you look closely, you'll see that the x-axis ticks don't line up _exactly_ with the edges of the histogram bar outlines. Also, the bin widths should be uniform, and because we rounded the values, if we were just to look at the tick labels, we would think the bins alternate between having a width of 24 and a width of 23. We can try to fix these issues.

```python
# make a histogram of the data array
min_val = 20
max_val = 350
val_range = max_val - min_val
num_bins = 8 
bin_width = val_range/num_bins
new_max = min_val+(bin_width*num_bins)+1 # Add one because range doesn't include upper value

# Since this plot has fewer bins, there is more space for each individual label
# This means we don't have to get rid of the decimals.
tick_values = list(getrange(min_val, new_max, bin_width))

plt.hist(data, bins=np.arange(min_val, new_max, bin_width),facecolor='blue', edgecolor = "k", alpha=0.3)
# data is the list of data values to make the histogram from
# bins uses the same values we used to create the tick_values list
# facecolor and opacity(alpha) set column appearance
# edgecolor = "k" outlines the columns in black

# make plot title & labels
plt.title('Histogram of Values in Dataset')
plt.xlabel("values")

plt.xticks(tick_values) # make upper limit 351 since range doesn't include upper bound
# range(lowest, highest+1, step value) 
# step value = bin width --> this ensures the bins are of equal width
# bin width = (highest - lowest) / number of bins --> since this might not be an integer value - cast it to an int
plt.ylabel("count") 

plt.show()

```

You could also look for a number of bins that evenly divides the range of the dataset you are working with (or modily the min and max a bit, as we did here by making the max value 350), so that you get an even value for your bin width w hen you divide val_range / num_bins.

```python
# make a histogram of the data array
min_val = 20
max_val = 350
val_range = max_val - min_val
num_bins = 15 # try different values until the plot looks acceptable to you 
bin_width = val_range/num_bins
new_max = min_val+bin_width*num_bins+1

tick_values_unrounded = list(getrange(min_val, new_max, bin_width))
tick_values = []
for val in tick_values_unrounded:
    tick_values.append(round(val))

plt.hist(data, bins=np.arange(min_val, new_max+1, bin_width),facecolor='blue', edgecolor = "k", alpha=0.3)
# data is the list of data values to make the histogram from
# bins uses the same values we used to create the tick_values list
# facecolor and opacity(alpha) set column appearance
# edgecolor = "k" outlines the columns in black

# make plot title & labels
plt.title('Histogram of Values in Dataset')
plt.xlabel("values")

plt.xticks(tick_values) # make upper limit 351 since range doesn't include upper bound
# range(lowest, highest+1, step value) 
# step value = bin width --> this ensures the bins are of equal width
# bin width = (highest - lowest) / number of bins --> since this might not be an integer value - cast it to an int
plt.ylabel("count") 

plt.show()
```

 This plot with 15 bins lines up perfectly because val_range/num_bins is an integer. The rounding step is actually unnecessary here. Even though the ticks line up with the histogram bar edges now, I am not sure I find the _shape_ of the plot as aesthetically pleasing.
